**Loading the iris dataset :**

```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt

iris = datasets.load_iris()
```

```python

target = iris.target #to access the labels for each record

for k in [0, 1, 2]:
    print("La classe : %s, nombre d'exemplaires: %s" % (k, len(target[target == k])))
```

**Observation des données :**

```python
data = iris.data  # For easier access to data
print(type(data), data.ndim, data.shape)
```

```python

fig = plt.figure(figsize=(10, 5))
fig.subplots_adjust(hspace=0.5, wspace=0.5)

clist = ['violet', 'yellow', 'blue'] # Colours for Species
colors = [clist[c] for c in iris.target]

# Graph 1: Length and width of sepals
ax1 = plt.subplot(1, 2, 1)
ax1.scatter(data[:, 0], data[:, 1], c=colors)
plt.xlabel('Longueur du sépal (cm)')
plt.ylabel('Largeur du sépal (cm)')

# Chart 2: Petal length and width
ax2 = plt.subplot(1, 2, 2)
ax2.scatter(data[:, 2], data[:, 3], c=colors)
plt.xlabel('Longueur du pétale (cm)')
plt.ylabel('Largeur du pétale (cm)')

# Added the legend
for ind, s in enumerate(iris.target_names):
    plt.scatter([], [], label=s, color=clist[ind])

plt.legend(scatterpoints=1, frameon=False, labelspacing=1, bbox_to_anchor=(1.8, 0.5), loc="center right", title="Espèces")

plt.show()
```

```python
import pandas as pd
import seaborn as sns

iris = datasets.load_iris()
df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
df['target'] = iris['target']
df['label'] = df['target'].apply(lambda x: iris['target_names'][int(x)])
sns.pairplot(df, hue='label', vars=iris['feature_names'], diag_kind='hist', height=2.5)

plt.show()
```
